```python
import requests , time 
import pandas as pd
from datetime import datetime
```

```python
def datetotime(date):
  time_tuple=date.timetuple()
  timestamp=round(time.mktime (time_tuple))
  return timestamp
```

```python
datetime.today()
```

```python
start=datetotime(datetime(2022,9,14))
end =datetotime(datetime.today())
end=datetotime(datetime(2022,9,15))

symbol="INFY"
```

```python
url = "https://priceapi.moneycontrol.com/techCharts/indianMarket/stock/history?symbol="+symbol+"&resolution=5&from="+str(start)+"&to=" +str(end)
```

```python
url
```

```python
resp =requests.get (url).json()
```

```python
data =pd.DataFrame(resp) 
```

```python
def timetodate(timestamp):
  return datetime.fromtimestamp(timestamp)
```

```python
data["date"] = data["t"].apply (lambda x : timetodate(x))
```

```python
data.drop (["t","s"],axis=1,inplace=True)
```

```python
data.rename(columns={'o':"open",'h':'high','l':'low','c':'close','v':'volume'},inplace=True)
```

```python
data.index=data["date"]
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

```python
dataset = data.copy()
dataset=dataset.tail(1)
```

```python
dataset
```

```python
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('close')
test_labels = test_features.pop('close')
```

```python
normalizer = tf.keras.layers.Normalization(axis=-1)
```

```python
open_ = np.array(train_features['open'])
```

```python
open_normalizer = layers.Normalization(input_shape=[1,], axis=None)
open_normalizer.adapt(open_)
```

```python
open_data=train_features['open']
```

```python
open_model = tf.keras.Sequential([
    open_normalizer,
    layers.Dense(units=1)
])

open_model.summary()
```

```python
open_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
%%time
history = open_model.fit(
    train_features['open'],
    train_labels,
    epochs=100,
    # Suppress logging.
    verbose=0,
    # Calculate validation results on 20% of the training data.
    validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [close]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
test_results = {}

test_results['hopen_model'] = open_model.evaluate(
    test_features['open'],
    test_labels, verbose=0)
```

Regression with a deep neural network (DNN)

```python
def build_and_compile_model(norm):
  model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
  return model
```

```python
dnn_open_model = build_and_compile_model(open_normalizer)
```
